```python
# Import our dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import and read in the data file
import codecs
f=codecs.open("sample_dataset.csv", "r", "ISO-8859-1")
f
```

```python
# Read in the data file
df1=pd.read_csv(f)
df1
```

```python
# Drop null values
df1.dropna()
```

```python
df1.columns
```

```python
df1.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df1.dtypes
```

```python
df1['Personal'] = pd.to_numeric(df1['Personal'])
df1['Business'] = pd.to_numeric(df1['Business'])
```

```python
df1.dtypes
```

```python
# Creating new columns for the percentage of type of Tourism
df1['Personal Percentage'] = df1.apply(lambda row: (row.Personal/(row.Personal+row.Business))*100, axis = 1) 
df1['Business Percentage'] = df1.apply(lambda row: (row.Business/(row.Personal+row.Business))*100, axis = 1) 
```

```python
df1.head()
```

```python
def pop_tourism(x):
    if x >= 50:
        return 'Personal'
    else:
        return 'Business'
```

```python
df1['Most Tourism'] = df1['Personal Percentage'].apply(pop_tourism)
df1.head()
```

```python
# Separate the Features (X) from the Target (y)
y = df1["Most Tourism"]
X = df1.drop(columns=["Most Tourism", 'Country', 'Year'])
```

```python
# Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=2, 
                                                    stratify=y)
X_train.shape
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
# Fit (train) or model using the training data
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Print the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
