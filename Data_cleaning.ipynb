```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
raw_inbound_df = pd.read_csv('Inbound_Tourism_Type_in_thousands.csv')
raw_inbound_df.head()
```

```python
# Clean inboud data
clean_inbound_df = raw_inbound_df[raw_inbound_df['Total'].str.contains('..')==False]
clean_inbound_df = clean_inbound_df.replace(to_replace='..',value=0)
clean_inbound_df['Country'] = clean_inbound_df['Country'].str.lower()
cols = ['Personal','Business','Total','Africa','Americas','East Asia and Pacific','Europe','Middle East','South Asia','Other not Classified']
clean_inbound_df[cols] = clean_inbound_df[cols].astype('float64', copy=False,errors='raise')
clean_inbound_df.head()
```

```python
# Check datatypes
clean_inbound_df.dtypes
```

```python
#Export to CSV
clean_inbound_df.to_csv('clean_inbound_data_thousands.csv')
```
