```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
raw_inbound_df = pd.read_csv('Inbound_Tourism_Type_in_thousands.csv')
raw_inbound_df.count()
```

```python
# Clean inboud data
clean_inbound_df = raw_inbound_df[raw_inbound_df['Total']!= '..']
clean_inbound_df = clean_inbound_df.replace(to_replace='..',value=0)
clean_inbound_df['Country'] = clean_inbound_df['Country'].str.lower()
cols = ['Personal','Business','Total','Africa','Americas','East Asia and Pacific','Europe','Middle East','South Asia','Other not Classified']
clean_inbound_df[cols] = clean_inbound_df[cols].astype('float64', copy=False,errors='raise')
clean_inbound_df.count()
```

```python
# Check datatypes
clean_inbound_df.dtypes
```

```python
#Export to CSV
clean_inbound_df.to_csv('clean_inbound_data_thousands.csv')
```

```python
# Read world indicator csv into dataframe
raw_world_ind_df = pd.read_csv('World_Indicators.csv')
raw_world_ind_df.head()
```

```python
#check for significant Null value columns
raw_world_ind_df.count()
```

```python
# eliminate significantly null value columns keeping columns with more than 2200 non Nan Values
clean_wolrd_ind_df = raw_world_ind_df.drop(columns=['Business Tax Rate','CO2 Emissions','Days to Start Business','Ease of Business','Energy Usage','Hours to do Tax','Lending Interest'])
clean_wolrd_ind_df.dtypes
```

```python
clean_wolrd_ind_df.head()
```

```python
# further cleaning of world ind data
clean_wolrd_ind_df['Country']=clean_wolrd_ind_df['Country'].str.lower()
cols = ['GDP','Health Exp/Capita','Tourism Inbound','Tourism Outbound']
clean_wolrd_ind_df[cols]=clean_wolrd_ind_df[cols].replace('[\$,]','',regex=True).astype(float)
clean_wolrd_ind_df['Year']=clean_wolrd_ind_df['Year'].str[-4:].astype(int)
clean_wolrd_ind_df.head()
```

```python
#write dataframe to csv
clean_wolrd_ind_df.to_csv('clean_world_ind.csv')
```
